```python
def word_divide_char(inputs):
    charecters = []
    for i in inputs:
        charecters.append(i)
    return charecters
```

```python
word_divide_char(x)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(tokenizer = word_divide_char)
```

```python
vectorizer
```

```python
x = vectorizer.fit_transform(x)
```

```python
x.shape
```

```python
vectorizer.vocabulary_
```

```python
print(x)
```

```python
x[0]
```

```python
feature_names = vectorizer.get_feature_names()
# get tfidf vector for first document
first_document_vector = x[0]
# print the scores
df = pd.DataFrame(first_document_vector.T.todense(),index = feature_names,columns = ["tfidf"])
df.sort_values(by = ["tfidf"],ascending = False)
```

```python
# logistic regressor
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.20,random_state = 42)
```

```python
log_class = LogisticRegression(penalty = 'l2',multi_class = 'ovr')
log_class.fit(x_train,y_train)
```

```python
print(log_class.score(x_test,y_test))
```

```python
# multinomial

clf = LogisticRegression(penalty = 'l2',multi_class = 'multinomial',solver = 'newton-cg')
clf.fit(x_train,y_train)
print(clf.score(x_test,y_test))
```

**Evaluation**

```python
x_predict = np.array(["abc"])
x_predict = vectorizer.transform(x_predict)
y_pred = log_class.predict(x_predict)
print(y_pred)
```
